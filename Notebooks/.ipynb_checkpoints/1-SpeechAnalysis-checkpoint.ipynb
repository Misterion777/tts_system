Trying to get the difference between different vowels in term of sound.

```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
```

```python
a_path = '../data/vowels/a.wav'
o_path = '../data/vowels/o.wav'
e_path = '../data/vowels/e.wav'
```

```python
def read(path):
    sample_rate, data = wavfile.read(path)
    return np.linspace(0,len(data),len(data)),data, sample_rate
```

```python
a_x,a_y, a_sr = read(a_path)
o_x,o_y, o_sr = read(o_path)
e_x,e_y, e_sr = read(e_path)
```

### Time domain

```python
plt.figure(figsize=(16, 6))

plt.subplot(131)
plt.title('a')
plt.plot(a_x[55000:75000],a_y[55000:75000])
plt.subplot(132)
plt.title('o')
plt.plot(o_x[65000:95000],o_y[65000:95000])
plt.subplot(133)
plt.title('e')
plt.plot(e_x[60000:80000],e_y[60000:80000])

plt.show()
```

```python
def spectrum(y,Fs):
    n = len(y) # length of the signal
    k = np.arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range
    frq = frq[range(n//2)] # one side frequency range

    Y = np.fft.fft(y)/n # fft computing and normalization
    Y = Y[range(n//2)]

    return frq[:Fs], np.abs(Y)[:Fs] # plotting the spectrum

    
# def spectrum(data,fs):
#     fourier = np.fft.fft(data)
#     freqs = np.fft.fftfreq(data.size)
#     return freqs,np.abs(fourier)
```

```python
a_freqs, a_m = spectrum(a_y,a_sr)
o_freqs, o_m = spectrum(o_y,o_sr)
e_freqs, e_m = spectrum(e_y,e_sr)
```

### Frequency domain

```python
plt.figure(figsize=(16, 6))

plt.subplot(131)
plt.title('a')
plt.plot(a_freqs,a_m)
plt.xlim(right=5000)

plt.subplot(132)
plt.title('o')
plt.plot(o_freqs,o_m)
plt.xlim(right=5000)

plt.subplot(133)
plt.title('e')
plt.plot(e_freqs,e_m)
plt.xlim(right=5000)

plt.show()
```
